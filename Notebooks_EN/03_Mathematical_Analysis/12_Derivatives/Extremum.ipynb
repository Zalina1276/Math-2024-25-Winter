## **Exercise 7**
### Largest Possible Rectangle with 10 Meters of String
Let:
- $ l $ be the length of the rectangle
- $ w $ be the width of the rectangle

The perimeter of the rectangle is given by the formula:

$P = 2l + 2w$

We are given that the perimeter is 10 meters:

$2l + 2w = 10$

This simplifies to:

$l + w = 5$

The area \( A \) of the rectangle is:

$A = l \cdot w$

We can solve for $ l $ in terms of $ w $:

$l = 5 - w$

Substitute this into the area formula:

$A(w) = (5 - w) \cdot w = 5w - w^2$

To maximize the area, we need to take the derivative of $ A(w) $

$A'(w) = \frac{d}{dw}(5w - w^2) = 5 - 2w$

Set the derivative equal to zero:

$5 - 2w = 0$

$2w=5$

$w=2.5$

From the perimeter equation $ l + w = 5 $,  we know:

$l = 5 - w = 5 - 2.5 = 2.5$

Dimensions of the rectangle are: length = 2.5 meters, width = 2.5 meters

## **Exercise 9**
Finding Extremum of the Function
$f(x) = \frac{x^2 + 2x + 1}{x - 1}$

$f'(x) = \frac{(2x + 2)(x - 1) - (x^2 + 2x + 1)(1)}{(x - 1)^2}$

Now, let's simplify the numerator:

$(2x + 2)(x - 1) = 2x(x - 1) + 2(x - 1) = 2x^2 - 2x + 2x - 2 = 2x^2 - 2$

So the numerator becomes:

$(2x^2 - 2) - (x^2 + 2x + 1) = 2x^2 - 2 - x^2 - 2x - 1 = x^2 - 2x - 3$

$f'(x) = \frac{x^2 - 2x - 3}{(x - 1)^2}$

To find the critical points, we set $f'(x) = 0 $:

$\frac{x^2 - 2x - 3}{(x - 1)^2} = 0$

$x^2 - 2x - 3 = 0$

$(x - 3)(x + 1) = 0$

$x = 3 \quad \text{or} \quad x = -1$

We found that $ x = 3 $ and $ x = -1 $ are critical points. However, the domain of the function has a restriction: $ x \neq 1 $ because the denominator $ (x - 1) $ cannot be zero.

- The critical points of $ f(x) = \frac{x^2 + 2x + 1}{x - 1} $ are $ x = 3 $ and $ x = -1 $
- The function has a vertical asymptote at $ x = 1$.
