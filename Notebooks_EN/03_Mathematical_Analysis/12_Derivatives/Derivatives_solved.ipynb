### 7. $ y(x) = \cos(x) + \sin(x) $

### 8. $ y(x) = 2\sin(x)\cos(x) $
This is a product of two functions, so we use the **product rule**. The product rule states that the derivative of $$ u(x)v(x)  =  u'(x)v(x) + u(x)v'(x) $$.

$y'(x) = 2\left( \cos(x) \cdot \cos(x) - \sin(x) \cdot \sin(x) \right) = 2\cos(2x)$

Explanation: Using the trigonometric identity $ 2\sin(x)\cos(x) = \sin(2x) $, we can simplify the derivative as $ 2\cos(2x) $.

### 11. $ y(x) = \frac{x}{x+1} $
We use the **quotient rule**, which states that for $ \frac{u(x)}{v(x)} $, the derivative is:

$$
y'(x) = \frac{v(x)u'(x) - u(x)v'(x)}{(v(x))^2}
$$
$y'(x) = \frac{(x+1)(1) - x(1)}{(x+1)^2} = \frac{1}{(x+1)^2}$

Explanation: The derivative of \( x \) is 1, and the derivative of \( x+1 \) is also 1. So the quotient rule gives us the result.

## **Exercise 2**
$\frac{d}{dx} \left(\ln(\sin(x))\right) = \cot(x)$

We know that:  

$$
\frac{d}{dx} \left(\ln(u(x))\right) = \frac{u'(x)}{u(x)}
$$

Let $ u(x) = \sin(x) $, so:  

$\frac{d}{dx} \left(\ln(\sin(x))\right) = \frac{\frac{d}{dx}(\sin(x))}{\sin(x)}$

The derivative of $ \sin(x) $ is:  

$\frac{d}{dx}(\sin(x)) = \cos(x)$

So we substitute this result back into the expression:

$\frac{d}{dx} \left(\ln(\sin(x))\right) = \frac{\cos(x)}{\sin(x)}$

From trigonometry, we know that:  

$$
\cot(x) = \frac{\cos(x)}{\sin(x)}
$$

Thus:  

$\frac{d}{dx} \left(\ln(\sin(x))\right) = \cot(x)$

- $\displaystyle \lim_{x \to 0} \frac{\sin(x)}{x} $

###  Check Indeterminate Form  
We substitute $ x = 0 $:  

$\frac{\sin(0)}{0} = \frac{0}{0}$  

This is an indeterminate form, so we can apply **de l'HÃ´pital's Rule**, which states:  

$$
\displaystyle \lim_{x \to c} \frac{f(x)}{g(x)} = \lim_{x \to c} \frac{f'(x)}{g'(x)}
$$  

$\displaystyle \lim_{x \to 0} \frac{\sin(x)}{x}= \frac{\cos(x)}{1} = \cos(0) = 1$  

---
- $\displaystyle \lim_{x \to \infty} \frac{\ln(x)}{x} $  

### Check Indeterminate Form  
We substitute $ x \to \infty $:  

$\frac{\ln(x)}{x} \to \frac{\infty}{\infty}$  

This is an indeterminate form, so we apply de l'HÃ´pital's Rule.

$f(x) = \ln(x), \quad f'(x) = \frac{1}{x}$  

$g(x) = x, \quad g'(x) = 1$  

Applying de l'HÃ´pital's Rule:

$\displaystyle \lim_{x \to \infty} \frac{\frac{1}{x}}{1} = \displaystyle \lim_{x \to \infty} \frac{1}{x}$  

As $ x \to \infty $, $ \frac{1}{x} \to 0 $

$\displaystyle \lim_{x \to \infty} \frac{\ln(x)}{x} = 0$

- $\displaystyle \lim_{x \to \infty} \frac{e^x}{x} $   

### Check Indeterminate Form  

$\frac{e^x}{x} \to \frac{\infty}{\infty}$  

This is an indeterminate form, so we apply de l'HÃ´pital's Rule.

$f(x) = e^x, \quad f'(x) = e^x$  

$g(x) = x, \quad g'(x) = 1$  

Applying de l'HÃ´pital's Rule:

$\displaystyle \lim_{x \to \infty} \frac{e^x}{1} = \displaystyle  \lim_{x \to \infty} e^x$  

Since $ e^x \to \infty $, the result is:

$ \displaystyle \lim_{x \to \infty} \frac{e^x}{x} = \infty$
