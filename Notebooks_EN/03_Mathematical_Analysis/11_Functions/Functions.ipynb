## **Exercise 3**
3. Let $f(x) = e^x$ and $g(x) = \ln(x)$. Check: $f(g(x))$ and $g(f(x))$

1.We substitute $g(x) = \ln(x) $ into $ f(x) $:  

$
f(g(x)) = f(\ln(x)) = e^{\ln(x)}
$  

Since:  

$
e^{\ln(x)} = x \quad \text{for } x > 0
$  

We get:  

$
f(g(x)) = x \quad \text{for } x > 0
$  

2.We substitute $ f(x) = e^x $ into $ g(x) $:  

$
g(f(x)) = g(e^x) = \ln(e^x)
$  

Since:  

$
\ln(e^x) = x \quad \text{for all real } x
$  

We get:  

$
g(f(x)) = x
$  
**Conclusion:**  
- $ f(g(x)) = x$ for $ x > 0 $
- $ g(f(x)) = x $ for all real $ x $

**What We Notice:**  
The functions $f(x) = e^x$ and $ g(x) = \ln(x) $ are **inverses** of each other. Composing them in either order returns the input, showing that they undo each other's effects. This is a key property of inverse functions

## **Exercise 6**

 We have function $f(x)= x-1$. Give inverse function $f^{-1}$.

1. Start with $y = f(x) = x - 1 $.
2. Swap $ x $ and $y $ to find the inverse:  
   $ x = y - 1 $

  $ y = x + 1 $

Thus, the inverse function is:

$
f^{-1}(x) = x + 1
$
