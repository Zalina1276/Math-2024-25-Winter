## **Exercise 1**
$ P=(1,2,3)$ , $ r=3$

The general equation of a sphere with center at $(h, k, l)$ and radius $r$ is:

$$ (x - h)^2 + (y - k)^2 + (z - l)^2 = r^2 $$

$ (x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 3^2 $

$ (x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9 $

### Final Answer:

The equation of the sphere is:

$$ (x - 1)^2 + (y - 2)^2 + (z - 3)^2 = 9 $$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parameters for the sphere
center = np.array([1, 2, 3])  # Center of the sphere
radius = 3  # Radius of the sphere

# Create a grid of points for the sphere
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = center[0] + radius * np.outer(np.cos(u), np.sin(v))
y = center[1] + radius * np.outer(np.sin(u), np.sin(v))
z = center[2] + radius * np.outer(np.ones(np.size(u)), np.cos(v))

# Create the figure and axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface of the sphere
ax.plot_surface(x, y, z, color='b', alpha=0.6)

# Label the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the title
ax.set_title('Sphere with Center (1, 2, 3) and Radius 3')

# Show the plot
plt.show()
```

### **Exercise 2**

1. $x^2 + y^2 + z^2 = 1$
2. $x^2 + y^2 + z^2 = 2$

$x^2 + y^2 + z^2 = r^2$

- The first sphere, $x^2 + y^2 + z^2 = 1$, has a radius $r_1 = 1$.
- The second sphere, $x^2 + y^2 + z^2 = 2$, has a radius $r_2 = \sqrt{2}$.

Both spheres have the same center at the origin, but different radius.

#### Relationship between the spheres
The two spheres have different radius, and since they are concentric (both centered at the origin), they do **not** intersect. Here's why:

- The sphere with the larger radius will encompass the sphere with the smaller radius entirely.
- For two concentric spheres to intersect, their radii must be equal. If the radii are not equal, the spheres will not intersect.

#### Conclusion
Since the two spheres have different radius, they do not have any common points.

Thus, **the spheres $x^2 + y^2 + z^2 = 1$ and $x^2 + y^2 + z^2 = 2$ do not intersect and have no common points**.

#### Geometric Interpretation
- The first sphere is entirely inside the second sphere.
- The two spheres do not overlap at any point.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to plot a sphere
def plot_sphere(ax, radius, center):
    u = np.linspace(0, 2 * np.pi, 100)
    v = np.linspace(0, np.pi, 100)
    x = center[0] + radius * np.outer(np.cos(u), np.sin(v))
    y = center[1] + radius * np.outer(np.sin(u), np.sin(v))
    z = center[2] + radius * np.outer(np.ones(np.size(u)), np.cos(v))

    ax.plot_surface(x, y, z, color='b', alpha=0.3)

# Create figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Sphere 1 (radius=1)
radius1 = 1
center1 = [0, 0, 0]  # Center of the sphere
plot_sphere(ax, radius1, center1)

# Sphere 2 (radius=sqrt(2))
radius2 = np.sqrt(2)
center2 = [0, 0, 0]  # Same center
plot_sphere(ax, radius2, center2)

# Setting plot limits
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Visualization of Spheres with $x^2 + y^2 + z^2 = 1$ and $x^2 + y^2 + z^2 = 2$')

# Show plot
plt.show()
```

## **Exercise 3*
The first sphere is:
$x^2 + y^2 + z^2 = 1$

The second sphere is:
$(x - 1)^2 + y^2 + z^2 = 1$

1.$(x^2 - 2x + 1) + y^2 + z^2 = 1$

$x^2 + y^2 + z^2 - 2x + 1 = 1$

$x^2 + y^2 + z^2 - 2x = 0$

2.$\left(x^2 + y^2 + z^2 - 2x\right) - (x^2 + y^2 + z^2) = 0 - 1$

$-2x = -1,$

$x = \frac{1}{2}$

3.$\left(\frac{1}{2}\right)^2 + y^2 + z^2 = 1$

$\frac{1}{4} + y^2 + z^2 = 1$

Subtract $\frac{1}{4}$ from both sides:

$y^2 + z^2 = \frac{3}{4}$

Thus, the equation of the curve formed by the intersection of the two spheres is:
$x = \frac{1}{2}, \quad y^2 + z^2 = \frac{3}{4}.$

This describes a **circle** in the plane $x = \frac{1}{2}$, with radius $\frac{\sqrt{3}}{2}$ in the $yz$ - plane.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the grid for the plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Sphere 1: x^2 + y^2 + z^2 = 1
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x1 = np.outer(np.cos(u), np.sin(v))
y1 = np.outer(np.sin(u), np.sin(v))
z1 = np.outer(np.ones(np.shape(u)), np.cos(v))

# Sphere 2: (x - 1)^2 + y^2 + z^2 = 1
x2 = 1 + np.outer(np.cos(u), np.sin(v))
y2 = np.outer(np.sin(u), np.sin(v))
z2 = np.outer(np.ones(np.shape(u)), np.cos(v))

# Plot the spheres
ax.plot_surface(x1, y1, z1, color='b', alpha=0.5)
ax.plot_surface(x2, y2, z2, color='r', alpha=0.5)

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title('Intersection of Two Spheres')
plt.show()
```

## **Exercise 4**
The equation of the tangent plane to the paraboloid $z = (x - 1)^2 + y^2 + 1$ at the point $P = (1, 0, 1)$

The paraboloid is given by:  
$F(x, y, z) = z - (x - 1)^2 - y^2 - 1 = 0.$

The gradient of $F(x, y, z)$ is:  
$\nabla F = \left( \frac{\partial F}{\partial x}, \frac{\partial F}{\partial y}, \frac{\partial F}{\partial z} \right) = \left( -2(x - 1), -2y, 1 \right).$

### the gradient at $P = (1, 0, 1)$  
At $P = (1, 0, 1)$,  

$\nabla F(1, 0, 1) = \left( -2(1 - 1), -2(0), 1 \right) = (0, 0, 1).$

The general form of a plane is:  
$$0 = \nabla F \cdot \left( (x, y, z) - P \right),$$  

Substituting $\nabla F = (0, 0, 1)$ and $P = (1, 0, 1)$, we get:

$0 = 0(x - 1) + 0(y - 0) + 1(z - 1).$

$z - 1 = 0.$

The equation of the tangent plane is:  
$z = 1.$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the paraboloid
x = np.linspace(-1, 3, 100)
y = np.linspace(-2, 2, 100)
x, y = np.meshgrid(x, y)
z_paraboloid = (x - 1)**2 + y**2 + 1

# Define the tangent plane
z_tangent_plane = np.ones_like(z_paraboloid)

# Create the 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the paraboloid
ax.plot_surface(x, y, z_paraboloid, alpha=0.7, rstride=10, cstride=10, cmap='viridis', edgecolor='none', label="Paraboloid")

# Plot the tangent plane
ax.plot_surface(x, y, z_tangent_plane, alpha=0.5, color='red', label="Tangent Plane")

# Plot the tangent point
ax.scatter(1, 0, 1, color='black', s=50, label="Tangent Point")

# Labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title("Paraboloid and Tangent Plane")

plt.legend()
plt.show()
```
