## **Exercise 7**

The formula for the distance $d$ from a point $(x_1, y_1, z_1)$ to a plane $Ax + By + Cz + D = 0$ is:

$$d = \frac{|Ax_1 + By_1 + Cz_1 + D|}{\sqrt{A^2 + B^2 + C^2}}$$

1)$2x + 3y + 4z - 5 = 0$

Here, $A = 2$, $B = 3$, $C = 4$, and $D = -5$. The point $A(1, 2, 3)$ has coordinates $(x_1, y_1, z_1) = (1, 2, 3)$.

$d = \frac{|2(1) + 3(2) + 4(3) - 5|}{\sqrt{2^2 + 3^2 + 4^2}}$

$d = \frac{|2 + 6 + 12 - 5|}{\sqrt{4 + 9 + 16}} = \frac{|15|}{\sqrt{29}} = \frac{15}{\sqrt{29}}$

## **Exercise 8**
The plane intersects the coordinate axes at points $A(2, 0, 0)$, $B(0, 3, 0)$, and $C(0, 0, 4)$

General formula of the plane equation:

$$Ax + By + Cz = D$$

1)For point $A(2, 0, 0)$:

$A(2) + B(0) + C(0) = D \implies 2A = D$

2)For point $B(0, 3, 0)$:

$A(0) + B(3) + C(0) = D \implies 3B = D$

3)For point $C(0, 0, 4)$:

$A(0) + B(0) + C(4) = D \implies 4C = D$

Now, we have the system of equations:

$\begin{cases}2A = D\\ 3B = D\\ 4C = D\end{cases}$

Thus, $A : B : C = 1 : \frac{1}{2} : \frac{1}{4}$

This ratio tells us that for any plane equation, the coefficients $A$, $B$, and $C$ should be in the ratio $1 : \frac{1}{2} : \frac{1}{4}$

The equation of the plane is:

$x + \frac{y}{2} + \frac{z}{4} = 1$

## **Exercise 9**
Plane $x + y + z = 1$ and the plane $x = 0$ (the $yz$ plane)

1)$\vec{n_1}=x + y + z = \langle 1, 1, 1 \rangle$.

$\vec{n_2}=x= \langle 1, 0, 0 \rangle$.

The formula to calculate the angle $\theta$ between two vectors $\vec{n_1}$ and $\vec{n_2}$ is:

$$ \cos \theta = \frac{\vec{n_1} \cdot \vec{n_2}}{\|\vec{n_1}\| \|\vec{n_2}\|} $$

Where:
- $\vec{n_1} \cdot \vec{n_2}$ is the dot product of the vectors.
- $\|\vec{n_1}\|$ and $\|\vec{n_2}\|$ are the magnitudes (norms) of the vectors.

2)$ \vec{n_1} \cdot \vec{n_2} = 1 \times 1 + 1 \times 0 + 1 \times 0 = 1 $

The magnitudes of $\vec{n_1}$ and $\vec{n_2}$ are:

3)$ \|\vec{n_1}\| = \sqrt{1^2 + 1^2 + 1^2} = \sqrt{3} $

$ \|\vec{n_2}\| = \sqrt{1^2 + 0^2 + 0^2} = 1 $

4)$ \cos \theta = \frac{1}{\sqrt{3} \times 1} = \frac{1}{\sqrt{3}} $

5)$ \theta = \cos^{-1}\left(\frac{1}{\sqrt{3}}\right) $
