## The slope-interception of two points:

$$$m = \frac{(y2 - y1)}{(x2 - x1)}$$

##The slope-intercept form of a line is given by the equation:

$$y=mx+b$$


where $m$ is the slope and is the y-intercept.

**Parallel lines have the same slope**

##The slope of a line perpendicular to another line is the negative reciprocal of the original line's slope:

$$𝑚_{perpendicular}=−\frac{1}{m_{given}}$$


## The point-slope form of a line (the equation of a line):

$$y-y1=m(x-x1)$$

## The interception point of two lines occurs where their equations are equal:

$$y_1=y_2$$

##The parametric form of the line passing through a point (𝑥1,𝑦1) and parallel to a vector $\vec{𝑣}=[𝑎,𝑏]$
$$x=𝑥1+𝑎𝑡$$

$$𝑦=𝑦1+𝑏𝑡$$

where
$𝑡$ is a real number parameter.

##To find the distance from a point $𝐴(𝑥1,𝑦1)$ to a line in the form $Ax+By+C=0$, we can use the following formula:

$$Distance=\frac{|A_{x1}​+B_{y1}+C|}{\sqrt{A^2 + B^2}}$$

##The angle $\theta $ between a line and the x-axis can be calculated using the formula:
$$
\tan(\theta) = |m|
$$

Now, to find $\theta $, take the arctangent (inverse tangent) of both sides:
$$
\theta = \tan^{-1}
$$

##The normal vector to a line in the form \( Ax + By + C = 0 \) is given by:
$$
\vec{n} = \langle A, B \rangle
$$

$x + y + 1 = 0$

$ A = 1,B = 1,C = 1

$\vec{n} = \langle A, B \rangle$ = $ \langle 1, 1 \rangle$

A vector perpendicular to the line will be parallel to the normal vector, so the vector $\langle 1, 1 \rangle$ is already perpendicular to the line.


```python
import sympy

# Define symbols
x, y = sympy.symbols('x y')

# Define the line equation
line_eq = sympy.Eq(y, 2*x + 3)

# Define the point
point_A = (1, 2)

# Rearrange the line equation to the form Ax + By + C = 0
line_eq_rearranged = sympy.Eq(2*x - y + 3, 0)

# Extract coefficients A, B, and C
A = line_eq_rearranged.lhs.coeff(x)
B = line_eq_rearranged.lhs.coeff(y)
C = line_eq_rearranged.lhs.as_expr().subs({x: 0, y: 0})

# Calculate the distance using the formula
distance = sympy.Abs(A*point_A[0] + B*point_A[1] + C) / sympy.sqrt(A**2 + B**2)

# Display the result
display(distance)
```
